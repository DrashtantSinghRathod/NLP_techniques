```python
syll=[]
def syl(w):
    curr1=""
    
    for i in range(len(w)-1):
        curr1+=w[i]
        if w[i]== '्' :
            curr1 = curr1[:-2]
            curr1+=w[i-1]+chr(2381)
            continue
        if w[i] in vowel and w[i]!='्' and not w[i].isdigit():
            syll.append(curr1)
            curr1=""
            continue
        if w[i] not in vowel  and not w[i].isdigit():
            if w[i+1] not in vowel:
                    syll.append(curr1)
                    curr1=""
                    continue
        
    if curr1!="":
        syll.append(curr1)
    l=len(w)
         
    if l!=1:    
        if w[l-1] in vowel and not w[l-1].isdigit():
            x=syll.pop()
            syll.append(x+w[l-1])
        
        if w[l-1] not in  vowel  and not w[l-1].isdigit():
            if w[i-2]!='्':
                syll.append(w[l-1])   
            if w[i-2]=='्':
                y=syll.pop()
                syll.append(y+w[l-1])
        if w[l-1].isdigit():
            z=syll.pop()
            syll.append(z+w[l-1])
        
            
    else:
        syll.append(w)
    
        
    return syll
```
